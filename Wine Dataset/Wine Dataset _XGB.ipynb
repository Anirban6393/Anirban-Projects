## 1. Download Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine

features, target = load_wine(return_X_y=True)
```

```python
features.shape
```

```python
target.shape
```

```python
features=pd.DataFrame(features)
target=pd.DataFrame(target)
l_wine=pd.concat([features, target],axis=1)
l_wine.columns=['feature_1','feature_2','feature_3','feature_4','feature_5','feature_6','feature_7','feature_8','feature_9','feature_10','feature_11','feature_12','feature_13','target']
l_wine
```

```python
l_wine.info()
```

```python
l_wine['target'].value_counts()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(features, target, random_state=8)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

## 2. Prediction With XGBoost Classifier

```python
from xgboost import XGBClassifier

xgb=XGBClassifier(n_estimators=100, learning_rate=0.12, seed=100)
xgb.fit(x_train, y_train)
```

```python
from sklearn.model_selection import cross_validate

# cross validate XGBoost Classifier
scores_xgb = cross_validate(xgb, x_train, y_train, cv=8, return_train_score=True,return_estimator=False)
scores_xgb =pd.DataFrame(scores_xgb)
scores_xgb
```

```python
print('Mean Test Score of XGBoost Classifier: ', scores_xgb ['test_score'].mean())
print('Mean Train Score of XGBoost Classifier: ', scores_xgb ['train_score'].mean())
```

```python
y_pred_xgb = xgb.predict(x_test)
y_pred_xgb
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns
ax = sns.heatmap(confusion_matrix(y_test, y_pred_xgb), annot=True, fmt='d')
ax.set(xlabel='XGB_Prediction', ylabel='Truth')
```

```python
print(classification_report(y_test, y_pred_xgb))
```

```python
accuracy_score(y_pred_xgb, y_test)
```
